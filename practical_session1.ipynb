Create a target function ùëì	and a bidimensional dataset ùê∑ and see how the perceptron learningalgorithm works. Choose a random line in the plane as your target function, where one side of theline maps to +1 and the other maps to -1. Choose the inputs ùë•! of the data set as random points in theplane and evaluate the target function on each ùë•! to get the corresponding output ùë¶!.

```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Generating x values from 0 to 100
x = np.linspace(0, 100, 100)

# Generating a random slope and intercept
slope = random.uniform(-10, 10)
intercept = random.uniform(-50, 50)

# Calculating y values based on the random slope and intercept
y = slope * x + intercept

# Creating the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'y = {slope:.2f}x + {intercept:.2f}')
plt.title('Random Slope and Intercept Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.ylim(-100, 100)  # Setting the y-axis limits to ensure visibility
plt.legend()
plt.grid(True)
plt.show()
```
